```python
class VideoClip():
    def __init__(self, name:str, folder='train') -> None:
        self.path = name
        self.image_paths = image_paths = glob(f"{self.path}\\**\\**.jpg")
        self.label_path = glob(f"{self.path}\\**.json")[0]

    def __len__(self):
        return len(self.image_paths)

    def play_clip(self)->None:
        image_folder = '.'
        video_name = 'mygeneratedvideo.avi'

        frame = cv2.imread(os.path.join(image_folder, self.image_paths[0])) 

        video = cv2.VideoWriter(video_name, 0, 1, (CFG.w, CFG.h))  

        # Appending the images to the video one by one 
        for image in self.image_paths:  
            video.write(cv2.imread(os.path.join(image_folder, image)))

        cap = cv2.VideoCapture(video_name)
        while(cap.isOpened()): 
            
        # Capture frame-by-frame 
            ret, frame = cap.read() 
            if ret == True: 
            # Display the resulting frame 
                cv2.imshow('Frame', frame) 
                
            # Press Q on keyboard to exit 
                if cv2.waitKey(25) & 0xFF == ord('q'): 
                    break
        
        # Break the loop 
            else: 
                break

        # Deallocating memories taken for window creation 
        cv2.destroyAllWindows()  
        video.release()

        os.remove(video_name)
```
